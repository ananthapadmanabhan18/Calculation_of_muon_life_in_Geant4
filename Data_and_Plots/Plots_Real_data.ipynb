# __Plots and Calculations for Muon Lifetime from REAL DATA__ 

```python
import numpy as np
```

## Check for operating voltage

```python

data_650_200 = np.loadtxt('Real_Data//650_200.data')
data_700_200 = np.loadtxt('Real_Data//700_200.data')
data_750_200 = np.loadtxt('Real_Data//750_200.data')


data_650_200 =  [data_650_200[:,0][i] for i in range(len(data_650_200[:,0])) if data_650_200[:,0][i] < 40000]
data_700_200 =  [data_700_200[:,0][i] for i in range(len(data_700_200[:,0])) if data_700_200[:,0][i] < 40000]
data_750_200 =  [data_750_200[:,0][i] for i in range(len(data_750_200[:,0])) if data_750_200[:,0][i] < 40000]


bins=60
import matplotlib.pyplot as plt
fig, ax = plt.subplots(3, 1, figsize=(7,11))

ax[0].hist(data_650_200, bins, alpha=0.5, label=r'$V_{OP} = 650 V$',color='r')
ax[1].hist(data_700_200, bins, alpha=0.5, label=r'$V_{OP} = 700 V$',color='g')
ax[2].hist(data_750_200, bins, alpha=0.5, label=r'$V_{OP} = 750 V$',color='b')

ax[0].set_xlabel('Time (ns)')
ax[0].set_ylabel('N(t)')
ax[0].title.set_text(r"$V_{op} = 650V$")
ax[0].grid()

ax[1].set_xlabel('Time (ns)')
ax[1].set_ylabel('N(t)')
ax[1].title.set_text(r"$V_{op} = 700V$")
ax[1].grid()

ax[2].set_xlabel('Time (ns)')
ax[2].set_ylabel('N(t)')
ax[2].title.set_text(r"$V_{op} = 750V$")
ax[2].grid()

#fitting with equation : Ae^{-bx} 
from scipy.optimize import curve_fit
def func(x, A, b):
    return A*np.exp(-b*x)



#650V
y_650, binEdges_650 = np.histogram(data_650_200, bins=bins)
bincenters_650 = 0.5*(binEdges_650[1:]+binEdges_650[:-1])
popt_650, pcov_650 = curve_fit(func, bincenters_650, y_650, p0=[1, 0.0001])

#700V
y_700, binEdges_700 = np.histogram(data_700_200, bins=bins)
bincenters_700 = 0.5*(binEdges_700[1:]+binEdges_700[:-1])
popt_700, pcov_700 = curve_fit(func, bincenters_700, y_700, p0=[1, 0.0001])

#750V
y_750, binEdges_750 = np.histogram(data_750_200, bins=bins)
bincenters_750 = 0.5*(binEdges_750[1:]+binEdges_750[:-1])
popt_750, pcov_750 = curve_fit(func, bincenters_750, y_750, p0=[1, 0.0001])



ax[0].plot(bincenters_650, func(bincenters_650, *popt_650), 'r--', label=f'B= {popt_650[1]:.2e} ')
ax[1].plot(bincenters_700, func(bincenters_700, *popt_700), 'g--', label=f'B= {popt_700[1]:.2e} ')
ax[2].plot(bincenters_750, func(bincenters_750, *popt_750), 'b--', label=f'B= {popt_750[1]:.2e} ')


ax[0].legend()
ax[1].legend()
ax[2].legend()

fig.suptitle(r"$V_{th} = 200mV$ for different Operating voltage")

plt.tight_layout()
plt.show()

'''
Printing the lifetime of muon
'''


print("The Mean Life time of muon with operating Voltage 650 V:",1e-3/popt_650[1])
print("The Mean Life time of muon with operating Voltage 700 V:",1e-3/popt_700[1])
print("The Mean Life time of muon with operating Voltage 750 V:",1e-3/popt_750[1])
```
